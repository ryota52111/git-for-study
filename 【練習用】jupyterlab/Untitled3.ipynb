```python
df_concat=pd.DataFrame()

for i in file_path:
        df_read_excel=pd.read_excel(i)
        print(df_read_excel.head(3))
```

```python
df_concat=pd.DataFrame()

for i in file_path:
    df_read_excel=pd.read_excel(i)
    df_concat=pd.concat([df_read_excel,df_concat])
```
